```python
import pandas as pd

data = {
    "Имя": ["Анна", "Иван", "Ольга"],
    "Возраст": [25, 30, 22]
}

df = pd.DataFrame(data)
df["Город"] = ["Москва", "СПб", "Казань"]

print(df)
```

```python
df["Страна"] = "Россия"
print(df)
```

```python
df["Возраст_в_месяцах"] = df["Возраст"] * 12
print(df)
```

```python
df = df.assign(Зарплата=[50000, 60000, 45000])
print(df)
```

```python
df["Категория возраста"] = df["Возраст"].apply(lambda x: "Молодой" if x < 30 else "Взрослый")
print(df)
```

```python
df.loc[3] = ["Петр", 40, "Новосибирск", "Россия", 480, 85000, "Взрослый"]
print(df)
```

```python
new_row = pd.DataFrame([{"Имя": "Сергей", "Возраст": 35, "Город": "Сочи", "Страна": "Россия"}])
df = pd.concat([df, new_row], ignore_index=True)
print(df)
```

```python
import pandas as pd

data = {
    "Имя": ["Анна", "Иван", "Ольга"],
    "Возраст": [25, 30, 22],
    "Город": ["Москва", "СПб", "Казань"],
    "Зарплата": [50000, 60000, 45000]
}

df = pd.DataFrame(data)

df = df.drop(columns=["Зарплата"])
print(df)
```

```python
df = df.drop(columns=['Возраст', 'Город'])
print(df)
```

```python
import pandas as pd

data = {
    "Имя": ["Анна", "Иван", "Ольга", "Петр", "Мария"],
    "Возраст": [25, 30, 22, 40, 35],
    "Город": ["Москва", "СПб", "Казань", "Новосибирск", "СПб"],
    "Зарплата": [50000, 60000, 45000, 70000, 65000]
}

df = pd.DataFrame(data)
print(df)
```

```python
df_filtered = df.query("Возраст > 30")
print(df_filtered)
```

```python
df_filtered = df.query("Город == 'СПб' and Зарплата > 60000")
print(df_filtered)
```

```python
min_age = 25
df_filtered = df.query("Возраст > @min_age")
print(df_filtered)
```

```python
df_filtered = df[df["Город"].isin(["Москва", "СПб"])]
print(df_filtered)
```

```python
df_filtered = df[~df["Город"].isin(["Москва", "СПб"])]
print(df_filtered)
```

```python
df_filtered = df[df["Возраст"].between(25, 35)]
print(df_filtered)
```

```python
df_filtered = df[df["Возраст"].between(25, 35, inclusive="neither")]
print(df_filtered)
```

```python
import pandas as pd

data = {
    "Имя": ["Анна", "Иван", "Ольга", "Петр", "Мария"],
    "Возраст": [25, 30, 22, 40, 35],
    "Зарплата": [50000, 60000, 45000, 70000, 65000]
}

df = pd.DataFrame(data)
print(df)
```

```python
df_sorted = df.sort_values(by="Возраст")
print(df_sorted)
```

```python
df_sorted = df.sort_values(by="Возраст", ascending=False)
print(df_sorted)
```

```python
df_sorted = df.sort_values(by=["Возраст", "Зарплата"], ascending=[True, False])
print(df_sorted)
```

```python
df.loc[5] = ["Елена", None, 55000]
df_sorted = df.sort_values(by="Возраст", na_position="first")
print(df_sorted)
```

```python
df_sorted = df.sort_index()
print(df_sorted)
```

```python
df_sorted = df.sort_index(ascending=False)
print(df_sorted)
```
