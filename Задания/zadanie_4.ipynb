```python
# 4.1 Добавляю столбец "Категория зарплаты"
def category(salary):
    if salary < 60000:
        return "Низкая"
    elif salary < 100000:
        return "Средняя"
    else:
        return "Высокая"

df1["Категория зарплаты"] = df1["Зарплата"].apply(category)
print("4.1 Столбец 'Категория зарплаты' добавлен:")
print(df1[["Имя", "Зарплата", "Категория зарплаты"]])
print()
```

```python
# 4.2 Добавляю одного сотрудника через .loc[]
df1.loc[21] = [21, "Антон", 32, "Разработчик", "IT", 85000, 6, "Средняя"]
print("4.2 Добавлен сотрудник Антон:")
print(df1.tail())
print()
```

```python
# 4.3 Добавляю двух сотрудников через pd.concat()
new_employees = pd.DataFrame([
    {"ID": 22, "Имя": "Максим", "Возраст": 29, "Должность": "Аналитик", "Отдел": "Маркетинг", "Зарплата": 72000, "Стаж": 4, "Категория зарплаты": "Средняя"},
    {"ID": 23, "Имя": "Наталья", "Возраст": 36, "Должность": "Менеджер", "Отдел": "Продажи", "Зарплата": 92000, "Стаж": 10, "Категория зарплаты": "Средняя"}
])
df1 = pd.concat([df1, new_employees], ignore_index=True)
print("4.3 Добавлены два сотрудника (Максим и Наталья):")
print(df1.tail())
```
