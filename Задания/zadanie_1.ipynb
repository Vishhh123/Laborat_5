```python
#1. Создание DataFrame
# 1.1 Из словаря
import pandas as pd
import numpy as np

data = {
    "ID": [1,2,3,4,5],
    "Имя": ["Иван","Ольга","Алексей","Мария","Сергей"],
    "Возраст": [25,30,40,35,28],
    "Должность": ["Инженер","Аналитик","Менеджер","Программист","Специалист"],
    "Отдел": ["IT","Маркетинг","Продажи","IT","HR"],
    "Зарплата": [60000,75000,90000,80000,50000],
    "Стаж": [2,5,15,7,3]
}
df1 = pd.DataFrame(data)
print(df1)
```

```python
# 1.2 Из списка словарей
data2 = [
    {"ID":1,"Имя":"Иван","Возраст":25,"Должность":"Инженер","Отдел":"IT","Зарплата":60000,"Стаж":2},
    {"ID":2,"Имя":"Ольга","Возраст":30,"Должность":"Аналитик","Отдел":"Маркетинг","Зарплата":75000,"Стаж":5}
]
df2 = pd.DataFrame(data2)
print(df2)
```

```python
# 1.3 Из NumPy (случайный возраст)
np.random.seed(1)
age = np.random.randint(20,61,(5,1))
df3 = pd.DataFrame(age, columns=["Возраст"])
print(df3)
```
