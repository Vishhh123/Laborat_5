```python
import pandas as pd
from datetime import datetime

df = pd.DataFrame(columns=["Пункт", "Номер", "Время"])

while True:
    print("\n1 - Добавить 2 - Найти 3 - Удалить 4 - Сохранить 5 - Загрузить 6 - Выход")
    cmd = input("Выбор: ")

    if cmd == "1":
        p = input("Пункт: ")
        n = input("Номер: ")
        t = input("Время (ЧЧ:ММ): ")
        df.loc[len(df)] = [p, n, t]
        df = df.sort_values("Пункт")
        print("Добавлено!")

    elif cmd == "2":
        t = input("Время (ЧЧ:ММ): ")
        res = df[df["Время"] > t]
        if res.empty:
            print("Нет поездов после", t)
        else:
            print(res)

    elif cmd == "3":
        col = input("Какую колонку удалить? (Пункт/Номер/Время): ")
        if col in df.columns:
            df = df.drop(columns=[col])
            print("Удалено!")
        else:
            print("Нет такой колонки")

    elif cmd == "4":
        df.to_parquet("trains.parquet", index=False)
        print("Сохранено!")

    elif cmd == "5":
        try:
            df = pd.read_parquet("trains.parquet")
            print("Загружено!")
        except:
            print("Файл не найден")

    elif cmd == "6":
        break
```
